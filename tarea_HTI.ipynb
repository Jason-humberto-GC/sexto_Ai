```python
#diapositiva 19 
import nltk
carpeta_nombre = "documentos\\"
archivo_nombre = "dos.txt"

with open(carpeta_nombre+archivo_nombre,"r") as archivo:
    texto=archivo.read()
    
tokens=nltk.word_tokenize(texto,"spanish")
texto_nltk=nltk.Text(tokens)
lista_palabras=["del","que","esto"]
texto_nltk.dispersion_plot(lista_palabras)  #grafico de dispersion de palabras

distribucion=nltk.FreqDist(texto_nltk)

lista_frecuencias=distribucion.most_common()
print(lista_frecuencias)
```

```python
#diapositiva 20
import nltk
carpeta_nombre = "documentos\\"
archivo_nombre = "dos.txt"

with open(carpeta_nombre+archivo_nombre,"r") as archivo:
    texto=archivo.read()
    
tokens=nltk.word_tokenize(texto,"spanish")
texto_nltk=nltk.Text(tokens)

distribucion=nltk.FreqDist(texto_nltk)
print("distribucion de la palabra apellido:")
print(distribucion["apellido"]) #frecuencia de la palabra apellido
```
